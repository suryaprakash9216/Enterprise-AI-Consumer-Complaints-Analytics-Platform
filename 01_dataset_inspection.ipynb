```python
import pandas as pd
from pathlib import Path

# Project Paths
PROJECT_PATH = Path(r"C:\Projects\Enterprise AI Analytics Platform for Customer Complaints")

RAW_FILE = PROJECT_PATH / "data" / "raw" / "complaints.csv"
OUTPUT_FILE = PROJECT_PATH / "data" / "processed" / "complaints_cleaned.csv"
```

```python
def clean_chunk(chunk):
    
    # Standardise column names
    chunk.columns = (
        chunk.columns
             .str.strip()
             .str.lower()
             .str.replace(" ", "_")
             .str.replace("-", "_")
             .str.replace("?", "", regex=False)
    )

    # Convert dates
    date_columns = [
        "date_received",
        "date_sent_to_company"
    ]

    for col in date_columns:
        chunk[col] = pd.to_datetime(chunk[col], errors="coerce")

    # Fill missing text values
    text_columns = [
        "consumer_complaint_narrative",
        "company_public_response",
        "tags"
    ]

    for col in text_columns:
        chunk[col] = chunk[col].fillna("Not Provided")

    # Remove duplicate Complaint IDs within the chunk
    chunk.drop_duplicates(subset="complaint_id", inplace=True)

    return chunk
```

```python
from pathlib import Path

PROJECT_PATH = Path(r"C:\Projects\Enterprise AI Analytics Platform for Customer Complaints")

RAW_FILE = PROJECT_PATH / "complaints.csv"

# Create processed folder if it doesn't exist
(PROJECT_PATH / "data" / "processed").mkdir(parents=True, exist_ok=True)

OUTPUT_FILE = PROJECT_PATH / "data" / "processed" / "complaints_cleaned.csv"
```

```python
print(RAW_FILE)
print(RAW_FILE.exists())

print(OUTPUT_FILE)
print(OUTPUT_FILE.parent.exists())
```

```python
import pandas as pd
import time

chunksize = 100000
first_chunk = True
total_rows = 0
chunk_number = 0

start_time = time.time()

for chunk in pd.read_csv(
    RAW_FILE,
    chunksize=chunksize,
    low_memory=False
):
    chunk_number += 1

    cleaned_chunk = clean_chunk(chunk)

    cleaned_chunk.to_csv(
        OUTPUT_FILE,
        mode="w" if first_chunk else "a",
        header=first_chunk,
        index=False
    )

    total_rows += len(cleaned_chunk)
    first_chunk = False

    print(f"Chunk {chunk_number} processed | Rows so far: {total_rows:,}")

end_time = time.time()

print("ETL Completed Successfully!")
print(f"Total rows processed: {total_rows:,}")
print(f"Time taken: {(end_time - start_time) / 60:.2f} minutes")
```

```python
import pandas as pd

chunksize = 100000

total_rows = 0
missing_counts = None
duplicate_rows = 0
duplicate_complaint_ids = 0
seen_ids = set()

for chunk_number, chunk in enumerate(pd.read_csv(OUTPUT_FILE, chunksize=chunksize), start=1):
    
    total_rows += len(chunk)

    # Missing values
    if missing_counts is None:
        missing_counts = chunk.isnull().sum()
    else:
        missing_counts += chunk.isnull().sum()

    # Duplicate full rows inside chunk
    duplicate_rows += chunk.duplicated().sum()

    # Duplicate complaint IDs across full file
    ids = chunk["complaint_id"].dropna().astype(str)

    duplicate_complaint_ids += ids.isin(seen_ids).sum()
    seen_ids.update(ids)

    print(f"Checked chunk {chunk_number} | Rows checked: {total_rows:,}")

missing_report = pd.DataFrame({
    "Missing Count": missing_counts,
    "Missing %": (missing_counts / total_rows) * 100
})

print("Total Rows:", total_rows)
print("Duplicate Rows:", duplicate_rows)
print("Duplicate Complaint IDs:", duplicate_complaint_ids)

missing_report.sort_values("Missing %", ascending=False)
```

```python
import pandas as pd
from pathlib import Path

MYSQL_READY_FILE = PROJECT_PATH / "data" / "processed" / "complaints_mysql_ready.csv"

chunksize = 100000
first_chunk = True
total_rows = 0
removed_rows = 0

for chunk in pd.read_csv(OUTPUT_FILE, chunksize=chunksize, low_memory=False):

    before = len(chunk)

    # remove rows with missing complaint_id
    chunk = chunk.dropna(subset=["complaint_id"])

    # convert complaint_id to integer
    chunk["complaint_id"] = chunk["complaint_id"].astype("int64")

    # convert date columns to MySQL format
    chunk["date_received"] = pd.to_datetime(chunk["date_received"], errors="coerce").dt.strftime("%Y-%m-%d %H:%M:%S")
    chunk["date_sent_to_company"] = pd.to_datetime(chunk["date_sent_to_company"], errors="coerce").dt.strftime("%Y-%m-%d %H:%M:%S")

    removed_rows += before - len(chunk)
    total_rows += len(chunk)

    chunk.to_csv(
        MYSQL_READY_FILE,
        mode="w" if first_chunk else "a",
        header=first_chunk,
        index=False
    )

    first_chunk = False

print("MySQL ready file created")
print("Rows kept:", total_rows)
print("Rows removed:", removed_rows)
```

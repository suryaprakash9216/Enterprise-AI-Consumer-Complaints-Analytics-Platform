```python
import torch
print(torch.__version__)

from sentence_transformers import SentenceTransformer
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
import sys
print(sys.executable)
```

```python
print(model)
```

```python
import pandas as pd
from pathlib import Path

PROJECT_PATH = Path(r"C:\Projects\Enterprise AI Analytics Platform for Customer Complaints")

EMBEDDING_FILE = PROJECT_PATH / "data" / "processed" / "embedding_sample_100k.csv"

df = pd.read_csv(EMBEDDING_FILE)

df.head()
```

```python
embeddings = model.encode(
    df["final_text_for_embeddings"].tolist(),
    batch_size=64,
    show_progress_bar=True,
    convert_to_numpy=True
)

print(embeddings.shape)
```

```python
import faiss
import numpy as np

embedding_dim = embeddings.shape[1]

index = faiss.IndexFlatL2(embedding_dim)

index.add(embeddings.astype("float32"))

print("Total vectors stored:", index.ntotal)
```

```python
(PROJECT_PATH / "vector_db").mkdir(parents=True, exist_ok=True)
```

```python
# Save FAISS index
faiss.write_index(
    index,
    str(PROJECT_PATH / "vector_db" / "complaints_faiss.index")
)

# Save metadata
df[["complaint_id", "final_text_for_embeddings"]].to_csv(
    PROJECT_PATH / "vector_db" / "complaints_metadata.csv",
    index=False
)

print("FAISS index and metadata saved successfully.")
```

```python
query = "customer says credit report has incorrect information"

query_embedding = model.encode([query], convert_to_numpy=True).astype("float32")

distances, indices = index.search(query_embedding, k=5)

for i, idx in enumerate(indices[0]):
    print(f"\nResult {i+1}")
    print("Complaint ID:", df.iloc[idx]["complaint_id"])
    print("Distance:", distances[0][i])
    print("Text:", df.iloc[idx]["final_text_for_embeddings"][:500])
```
